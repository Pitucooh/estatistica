# Questão 01 - Análise de uma distribuição normal (2,5 pontos)

O consumo anual per capta de sorvete (em libras) nos Estados Unidos pode ser aproximado por uma distribuição normal, como pode ser visto na figura. (*Adaptado de: U.S. Department of Agriculture.*).

image.png

**a)** (0,5) Construa a distribuição normal do consumo anual per capta de sorvetes dos EUA, de acordo com o enunciado.

```python
# Carcaterização da população
mu, sigma = 17.9, 4.4

# Geração da distribuição normal das distãncias de frenagem
consumo_sorvetes = st.norm(loc = mu, scale = sigma)
```

**b)** (0,5) Qual consumo per capta de sorvetes representa o 95º percentil?

```python
consumo = consumo_sorvetes.ppf(0.95)
print(f'Consumo per capta de sorvetes do 95º percentil = {consumo:.2f} lb')
```

**c)** (0,5) Qual consumo per capta representa o terceiro quartil?

```python
consumo = consumo_sorvetes.ppf(0.75)
print(f'Consumo per capta de sorvetes do terceiro quartil = {consumo:.2f} lb')
```

**d)** (0,5) Qual é o menor consumo per capta de sorvetes que pode estar nos 10% maiores consumos per capta de sorvetes?

```python
consumo = consumo_sorvetes.ppf(0.9)
print(f'Menor consumo per capta de sorvetes entre os 10% maiores consumos per capta = {consumo:.2f} lb')
```

**e)** (0,5) Qual é o maior consumo per capta de um norte americano que pode estar nos menores 5% consumos per capta?

```python
consumo = consumo_sorvetes.ppf(0.05)
print(f'Maior consumo per capta de sorvete entre os 5% menores consumos per capta = {consumo:.2f} lb')
```

# Questão 02 - Distribuição amostral das médias (2,5 pontos)

Uma máquina está regulada para encher caixas de leite com um desvio padrão de 91 g. Uma amostra aleatória de 40 caixas tem uma média de 1,82 kg. Nesse cenário:

**a)** (0,75) Construa a distribuição amostral dos valores médios das massas das caixas de leite.



```python
# Carcaterização da população
mu, sigma = 1.82, 0.091

# Geração da distribuição das médias amostrais
amostras = 40
mu_x, std_x = mu, sigma/(amostras**(1/2))
cx_leite = st.norm(mu_x, std_x)
```


**b)** (0,5) Qual é a probabilidade de que a máquina preencha uma caixa de leite com menos de 1,8 kg?

```python
prob = cx_leite.cdf(1.8)
print(f'P(valor médio < 1,8) = {prob*100:.2f}%')
```


**c)** (0,5) Qual é a probabilidade de que a máquina preencha uma caixa de leite com mais de 1,85 kg?

```python
prob = cx_leite.sf(1.85)
print(f'P(valor médio > 1,85) = {prob*100:.2f}%')
```


**d)** (0,75) Qual é a probabilidade de que a máquina preencha uma caixa de leite com massa entre 1,75 e 1,85 kg?

```python
# Cálculo da probabilidade do item (c)
prob = cx_leite.cdf(1.85) - cx_leite.cdf(1.75)
print(f'P(1,75 < valor médio < 1,85) = {prob*100:.2f}%')
```

# Questão 03 - Intervalo de Confiança para a Média Populacional (2,5 pontos)

As distâncias percorridas até o trabalho (em milhas) de 30 pessoas estão apresentadas a seguir.

image.png

A célula abaixo registra essas informações na variável `dados`.

```python
# Conjunto de dados
dados = pd.Series([12, 9, 7, 2, 8, 7, 3, 27, 21, 10, 13, 7, 2, 30, 7,
                   6, 13, 6, 4, 1, 10, 3, 13, 6, 2, 9, 2, 12, 16, 18])
```

**a)** (0,5) Suponha que o desvio padrão populacional $\sigma$ é de 8 milhas. Nessa condição, preencha o código da célula abaixo com as informações pertinentes ao problema.

```python
# Estimativa pontual
x_barra = dados.mean()

# Desvio padrão populacional
sigma = 8
n_amostras = 30
sigma_x = sigma / (n_amostras ** (1/2))

# Amostras
amostras = st.norm(loc = x_barra, scale = sigma)
```

**b)** (0,75) Construa um intervalo de 95% de confiança para a média populacional $\mu$.

```python
# Nível de confiança
confianca = 0.95

# Intervalo de confiança
IC1 = st.norm.interval(confidence = confianca, loc = x_barra, scale = sigma_x)
print(f'Sigma conhecido --> IC para média = ({IC1[0]:.2f}, {IC1[1]:.2f}) milhas')
```

**c)** (0,75) Refaça o item anterior, admitindo que o desvio populacional $\sigma$ não é conhecido.

```python
std_amostral = dados.std(ddof = 1)
IC2 = st.t.interval(confidence = confianca, df = n_amostras - 1, loc = x_barra, scale = std_amostral/(n_amostras**(1/2)))
print(f'Sigma desconhecido --> IC para média = ({IC2[0]:.2f}, {IC2[1]:.2f}) milhas')
```

**d)** (0,5) Execute a célula a seguir e explique a diferença entre os comprimentos dos intervalos de confiança dos itens anteriores.

```python
# Representação gráfica
plt.figure(figsize=(8,2))
plt.hlines(0.5, IC1[0], IC1[1], colors='b')
plt.hlines(1.0, IC2[0], IC2[1], colors='r')
ax = plt.gca()
ax.get_yaxis().set_visible(False)
plt.ylim(0, 1.75)
plt.vlines(x_barra, 0, 1.75, colors='k', linestyles='dashed')
plt.title('Intervalos de Confiança 95%')
plt.xlabel('Distâncias percorridas até o trabalho (em milhas)')
plt.legend(['sigma conhecido', 'sigma desconhecido'], loc='upper left')
plt.show()
```

```python
#@title Resposta
Resp_Q3_d = 'O ICs com sigma conhecido tendem a ser mais curtos e precisos se comparados com sigmas desconhecidos devido a incerteza do desvio padrao amostral' #@param {type:"string"}
display(Resp_Q3_d)
```

# Questão 04 - Intervalo de Confiança para a Variância e Desvio Padrão (2,5 pontos)

O conjunto de dados representa o tempo (em minutos) gasto verificando e-mail para uma amostra aleatória de funcionários de uma empresa.

image.png

```python
# Conjunto de dados
tempo = pd.Series([7.5, 2, 12.1, 8.8, 9.4, 7.3, 1.9, 2.8, 7, 7.3])
```

**a)** (1,0) Complete a célula a seguir com as características da distribuição dos tempos gastos para verificar e-mail e a correspondente distribuição chi-quadrado.

```python
# Caracterização da distribuição
amostras = len(tempo)
s_amostral = tempo.std(ddof = 1)

# Distribuição chi-quadrado
distchi2 = st.chi2(df = amostras - 1)
```

**b)** (1,0) Use um nível de confiança de 98% e construa um intervalo de confiança para a variância populacional.

```python
# Nível de confiança
confianca = 0.98

# Determinação dos valores críticos
x2L = distchi2.ppf((1-confianca)/2)
x2R = distchi2.ppf((1+confianca)/2)

# Extremos do IC para variância
inf_var = (amostras-1)*s_amostral**2/x2R
sup_var = (amostras-1)*s_amostral**2/x2L

# Exibição do resultado
print(f'IC para a variância = ({inf_var:.2f}, {sup_var:.2f}) minutos^2')
```

**c)** (0,5) Use um nível de confiança de 98% e construa um intervalo de confiança para a desvio padrão populacional.

```python
# Extremos do IC para o desvio padrão
inf_std = ((amostras-1)*s_amostral**2/x2R)**0.5
sup_std = ((amostras-1)*s_amostral**2/x2L)**0.5

# Exibição do resultado
print(f'IC para o desvio padrão = ({inf_std:.2f}, {sup_std:.2f}) minutos')
```
